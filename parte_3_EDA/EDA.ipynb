### Estrutura dos dados
* Title - título do artigo no card de extração
* Subtitle - subtítulo do artigo no card de extração
* Image (yes/no)- indica se o artigo possui uma imagem de preview no card de extração
* Author - autor do artigo
* Publication - título da publicação à qual o artigo está vinculada, para o caso de artigos independentes está marcado como "No * * publication"
* Year - Month - Day - data em que o artigo foi publicado
* Tag - Tag à qual o artigo foi vinculado, lembrando que um artigo pode ser vinculado à até 5 tags
* Reading Time- tempo de leitura do artigo
* Claps - número de aplausos que o artigo recebeu até a data de coleta dos dados
* Comment (yes/no) - indica se a entrda é um comentário em um outro artigo
* Story Url - link para o artigo
* Author URL - link para a homepage do autor

### Top 10 artigos com mais aplausos

```python
def graph_top10(df, group_var, agg_var, agg_type, title="Top 10", xlabel="Não definido", ylabel="Não definido"):
    if agg_var == "Author":
        unique_author = df.drop_duplicates(subset=[group_var, agg_var])
        df = unique_author.copy()
    else:
        pass
    
    if agg_type == 'sum':
        group = pd.DataFrame(df.groupby([group_var])[agg_var].sum()).reset_index().sort_values(agg_var, ascending=False)
    elif agg_type == 'count':
        group = pd.DataFrame(df.groupby([group_var])[agg_var].count()).reset_index().sort_values(agg_var, ascending=False)
    
    if group_var == "Publication":
        top10 = group[group[group_var] != 'No publication'].head(10).sort_values(agg_var)
    else:
        top10 = group.head(10).sort_values(agg_var)

    top10_list = top10[group_var].tolist()
    top10_count_list = top10[agg_var].tolist()
    
    plt.figure(figsize=(11,6))

    plt.barh(top10_list, top10_count_list, color="mediumseagreen")
    for index, value in enumerate(top10_count_list):
        plt.text(value+(top10[agg_var].max()*0.005), index-0.2, str(f'{int(value):,}').replace(",", "."))#, fontweight='bold'

    plt.title(title, fontsize=14)
 
    plt.xlim(0,(top10[agg_var].max()+(top10[agg_var].max()*0.1)))
    plt.ylabel(ylabel)
    plt.xlabel(xlabel)
    plt.xticks(fontsize=10)

    plt.show()
```

### Autores com mais aplausos acumulados

```python
def graph_top10_autor_claps(df, agg_type=''):
    copy_medium = df.copy()
    copy_medium["count_artigos"] = 1
    sum_author = copy_medium.groupby("Author").sum()
    if agg_type == 'mean':
        sum_author["CA_Ratio"] = sum_author['Claps']/sum_author["count_artigos"]
    elif agg_type == 'sum':
        sum_author["CA_Ratio"] = sum_author['Claps']
    else:
        raise Exception("In agg_type parameter, you must choose 'mean' or 'sum'")
        
    sum_author['CAR_artigos'] = sum_author.CA_Ratio.astype(int).apply('{:,}'.format).astype(str) + '\n('+ sum_author.count_artigos.astype(str) +' artigos)'
    sum_author = sum_author.reset_index().sort_values('CA_Ratio', ascending=False)
    #sum_author
    top10_CA_Ratio = sum_author.head(10).sort_values('CA_Ratio')
    CA_Ratio_list = top10_CA_Ratio['Author'].tolist()
    count_CAR_list = top10_CA_Ratio["CA_Ratio"].tolist()
    count_CAR_art_list = top10_CA_Ratio["CAR_artigos"].tolist()

    plt.figure(figsize=(11,6))
    plt.barh(CA_Ratio_list, count_CAR_list, color="mediumseagreen")
    for index, value in enumerate(count_CAR_art_list):
        plt.text(int(value.split()[0].replace(",", ""))+(top10_CA_Ratio.CA_Ratio.max()*0.005), index-0.3, str(value).replace(",", "."))#, fontweight='bold'

    if agg_type == 'mean':
        plt.title("Top 10 autores com maiores médias de aplausos por artigo", fontsize=14)
    else:
        plt.title("Top 10 autores com mais aplausos acumulados", fontsize=14)        
    plt.xlim(0,(top10_CA_Ratio.CA_Ratio.max()+(top10_CA_Ratio.CA_Ratio.max()*0.1)))
    plt.ylabel('Autor')
    if agg_type == 'mean':
        plt.xlabel("Média de aplausos por artigo")
    else:
        plt.xlabel("Total de aplausos")
    plt.xticks(fontsize=10)

    plt.show()    
```

```python
graph_top10_autor_claps(medium_2, agg_type='sum')
```
